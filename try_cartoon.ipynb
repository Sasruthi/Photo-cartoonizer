```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_file(file):
    img=cv2.imread(file)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
    return img
```

```python
#getting the edgeline of the image
def edge(img,line_size,blur_value):
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    blur=cv2.medianBlur(gray,blur_value)
    edges=cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,line_size,blur_value)
    #plt.title(' EDGE LINE OF THE IMAGE')
    #plt.imshow(edges)
    return edges

```

```python
#get_input()
file=input("Enter the name of the Image file(format to be specified)\n")
img=read_file(file)
plt.imshow(edge(img,17,7))
```

```python
#reducing color palette
def color_quantization(img,k):#process that reduces the number of distinct colors used in an image
    #transform
    data=np.float32(img).reshape((-1,3))
    criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,20,0.001)
    ret,label,center=cv2.kmeans(data,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
    center=np.uint8(center)
    result=center[label.flatten()]
    result=result.reshape(img.shape)
    return result
```

```python
#get_input()
file=input("Enter the name of the Image file(format to be specified)\n")
img=read_file(file)
plt.imshow(color_quantization(img,3))
```

```python
#reduce noise
def blur():
    blurr=cv2.bilateralFilter(img,d=10,sigmaColor=200,sigmaSpace=200)
    plt.imshow(blurr)
#get_input()
file=input("Enter the name of the Image file(format to be specified)\n")
img=read_file(file)
blur()
```
